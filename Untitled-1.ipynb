```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('univariate_prophet_alex_o.csv')
df2 = pd.read_csv('multivariate_prophet_alex_o.csv')
df3 = pd.read_csv('arima_alex_o.csv')
df4 = pd.read_csv('autoarima_alex_o.csv')
df5 = pd.read_csv('xgboost_alex_o.csv')
```

```python
dfs = [df1, df2, df3, df4, df5]
model_names = ['Prophet (Univariate)', 'Prophet (Multivariate)', 'ARIMA', 'AutoARIMA', 'XGBoost']

concatenated_df = pd.concat([df.assign(Model=name) for df, name in zip(dfs, model_names)], keys=model_names)

# Calculate average performance for each model
avg_performance = concatenated_df.groupby('Model').mean().mean(axis=1)

# Set the background style to whitegrid
sns.set_style("whitegrid")

# Set the color palette for better visibility
sns.set_palette("pastel")

# Create a color palette for each model
colors = sns.color_palette("pastel", n_colors=len(model_names))

# Plotting RMSE, MAE, and MASE using a bar plot without lines
plt.figure(figsize=(15, 5))

# RMSE
plt.subplot(1, 3, 1)
rmse_sorted_models = concatenated_df.groupby('Model')['rmse'].mean().sort_values(ascending=False).index
sns.barplot(x='Model', y='rmse', data=concatenated_df, order=rmse_sorted_models, palette=colors, ci=None)
plt.title('RMSE')
plt.xlabel('Model')
plt.ylabel('RMSE')
plt.xticks(rotation=45, ha='right')  # Tilt the x-axis labels

# MAE
plt.subplot(1, 3, 2)
mae_sorted_models = concatenated_df.groupby('Model')['mae'].mean().sort_values(ascending=False).index
sns.barplot(x='Model', y='mae', data=concatenated_df, order=mae_sorted_models, palette=colors, ci=None)
plt.title('MAE')
plt.xlabel('Model')
plt.ylabel('MAE')
plt.xticks(rotation=45, ha='right')  # Tilt the x-axis labels

# MASE
plt.subplot(1, 3, 3)
mase_sorted_models = concatenated_df.groupby('Model')['mase'].mean().sort_values(ascending=False).index
sns.barplot(x='Model', y='mase', data=concatenated_df, order=mase_sorted_models, palette=colors, ci=None)
plt.title('MASE')
plt.xlabel('Model')
plt.ylabel('MASE')
plt.xticks(rotation=45, ha='right')  # Tilt the x-axis labels

# Improve layout
plt.tight_layout()

# Add a title for the entire figure

# Show the plot
plt.show()
```
